```python
# ステップ関数のグラフ
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    return np.array(x > 0, dtype=int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) # y軸の範囲を指定
plt.show()
```

```python
# シグモイド関数の実装
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.array([-1.0, 1.0, 2.0])
sigmoid(x)
```

```python
t = np.array([1.0, 2.0, 3.0])
1.0 + t
```

```python
# グラフの描画
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# シグモイド関数とステップ関数の比較
# 非線形関数
```

```python
# ReLU関数
def relu(x):
    return np.maximum(0, x)
```

```python
# グラフの描画
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.show()
```
