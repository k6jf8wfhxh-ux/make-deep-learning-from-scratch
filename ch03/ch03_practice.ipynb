```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    return np.array(x > 0, dtype=int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# シグモイド関数の実装
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.array([-1.0, 1.0, 2.0])
sigmoid(x)
```
