```python
# パーセプトロン
# パーセプトロンとは
# 単純な論理回路
# AND回路
# NANDゲートとORゲート
```

```python
# パーセプトロンの実装
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    else:
        return 1

print(AND(0, 0))
print(AND(1, 0))
print(AND(0, 1))
print(AND(1, 1))
```

```python
# 重みとバイアスによる実装
# ANDゲート
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(x*w) + b

    if tmp <= 0:
        return 0
    else:
        return 1

print(AND(1, 1))
print(AND(1, 0))
print(AND(0, 1))
print(AND(0, 0))
```

```python
# ORゲート
def OR(x1, x2):
    x = np.array([x1 ,x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b

    if tmp <= 0:
        return 0
    else:
        return 1

print(OR(1, 1))
print(OR(1, 0))
print(OR(0, 1))
print(OR(0, 0))
```

```python
# NANDゲート
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w*x) + b

    if tmp <= 0:
        return 0
    else:
        return 1

print(NAND(1, 1))
print(NAND(1, 0))
print(NAND(0, 1))
print(NAND(0, 0))
```

```python
# XORゲート
# 線形と非線形
# 多層パーセプトロン
# 既存ゲートの組み合わせ
```

```python
# XORゲートの実装
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y

print(XOR(1, 1))
print(XOR(1, 0))
print(XOR(0, 1))
print(XOR(0, 0))
```
